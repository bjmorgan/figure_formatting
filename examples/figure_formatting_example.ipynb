```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
from figure_formatting import figure_formatting as ff
ff.set_formatting() # set default formatting.
```

```python
plt.figure()
x = np.arange(0, 6, 0.1)
plt.plot(x, np.sin(x), label='$\sin(x)$')
plt.plot(x, np.cos(x), label='$\cos(x)$')
xlbl = plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.legend()
plt.title("(a) example plot of $f(x)$")
plt.savefig('sincosx.pdf', dpi=300)
plt.show()
```

```python
x = np.arange(7)
heights = np.array([0.1, 0.1, 0.4, 0.3, 0.1, 0.01, 0.0])
plt.figure(figsize=(1.25, 1.25))
plt.bar(x, heights, width=0.5, color=[ff.tableauP['red'],
                                      ff.tableauP['red'],
                                      ff.tableauP['orange'],
                                      ff.tableauP['blue'],
                                      ff.tableauP['green'],
                                      ff.tableauP['brown']],
        edgecolor=ff.nearly_black,
        linewidth=ff.linewidth,
        tick_label=[0,1,2,3,4,5,6])
plt.xlabel("O-Li$_{6-x}$Mn$_x$ octahedra")
plt.ylabel("fractions")
plt.text(3.85, 0.52, "cycled")
plt.ylim(0, 0.6)
plt.Line2D([0.0, 1.0], [0.0, 1.0], linewidth=ff.linewidth)
plt.savefig('OLiMn_octahedra.pdf', dpi=300)
plt.show()
```
